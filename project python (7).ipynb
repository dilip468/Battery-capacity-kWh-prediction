```python
df['Car full name'] = df['Car full name'].str.replace(' ', '', regex=False)
```

```python
df['Car full name']=df['Car full name'].str.replace('-', '', regex=False)
```

```python
df['Model']=df['Model'].str.replace(' ', '', regex=False)
df['Model']=df['Model'].str.replace('-', '', regex=False)
```

```python
df['Type of brakes']=df['Type of brakes'].str.replace('+', '', regex=False)
df['Type of brakes']=df['Type of brakes'].str.replace(')', '', regex=False)
df['Type of brakes']=df['Type of brakes'].str.replace('(', '', regex=False)
df['Type of brakes']=df['Type of brakes'].str.replace(' ', '', regex=False)
df
```

```python
df.dropna(inplace=True)
```

```python
df_encoded = pd.get_dummies(df, columns=['Car full name','Make','Model','Type of brakes','Drive type'], drop_first=True)
```

```python
df=df_encoded.astype('int32')
```

```python
x=df.drop('Battery capacity [kWh]',axis=1)
y=df['Battery capacity [kWh]']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np


# Initialize the scaler
scaler = MinMaxScaler()


scaler.fit(X_train.values.reshape(-1, 1))
scaled_data = scaler.transform(X_train.values.reshape(-1, 1))

# Output
print("Original data (X_test):\n",X_train)
print("Scaled data (Min-Max):\n", scaled_data)
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


# Step 2: 

# Step 3: Initialize and train the Decision Tree model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Step 4: Make predictions
y_pred = model.predict(X_test)

# Step 5: Compare actual vs predicted
result_df = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred
})

# Step 6: Print comparison DataFrame and accuracy
print("Comparison of Actual vs Predicted:\n")
print(result_df)

# Step 7: Accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f"\nModel Accuracy: {accuracy:.2f}")
```
